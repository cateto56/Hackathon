```python
# cleaning up income_poverty

def impute_income(cols):
    income = cols[0]
    age = cols[1]
    education = cols[2]
    
    if pd.isnull(income):

        if(education=="< 12 Years"):
            if (age in ["18 - 34 Years","35 - 44 Years"]):
                return "Below Poverty"
        
            else:
                return "<= $75,000, Above Poverty"
            
        elif(education=='College Graduate'):
            if(age not in ["18 - 34 Years","65+ Years"]):
                return "> $75,000"
            else:
                return "<= $75,000, Above Poverty"

        else:
            return "<= $75,000, Above Poverty"
        
    else:
        return income
obj_X_train['income_poverty'] = obj_X_train[['income_poverty','age_group','education']].apply(impute_income,axis=1)
obj_X_test['income_poverty'] = obj_X_test[['income_poverty','age_group','education']].apply(impute_income,axis=1)
#obj_X_train
```

```python
# cleaning up marital_status

def impute_marital(cols):
    status = cols[0]
    age = cols[1]
    education = cols[2]
    
    if pd.isnull(status):

        if (education=="12 Years" or education=="< 12 Years" or education=="Some College"):
            if(age=="18 - 34 Years" or age=="65+ Years"):
                return "Not Married"
            
            else:
                return "Married"

        else:
            return "Married"
    else:
        return status
obj_X_train['marital_status'] = obj_X_train[['marital_status','age_group','education']].apply(impute_marital,axis=1)
obj_X_test['marital_status'] = obj_X_test[['marital_status','age_group','education']].apply(impute_marital,axis=1)
#obj_X_train
```

```python
# cleaning up employment_status

def impute_status(cols):
    status = cols[0]
    age = cols[1]
    
    if pd.isnull(status):
        if(age=="65+ Years"):
            return "Not in Labor Force"

        else:
            return "Employed"
    else:
        return status
obj_X_train['employment_status'] = obj_X_train[['employment_status','age_group']].apply(impute_status,axis=1)
obj_X_test['employment_status'] = obj_X_test[['employment_status','age_group','education']].apply(impute_status,axis=1)
#obj_X_train
#obj_X_train[obj_X_train.isnull().any(axis=1)]
#obj_X_train
```

Label Encoding

```python

obj_X_train['sex'] = pd.get_dummies(obj_X_train['sex'],drop_first=True).astype(float)
obj_X_test['sex'] = pd.get_dummies(obj_X_test['sex'],drop_first=True).astype(float)

obj_X_train['marital_status'] = pd.get_dummies(obj_X_train['marital_status'],drop_first=True).astype(float)
obj_X_test['marital_status'] = pd.get_dummies(obj_X_test['marital_status'],drop_first=True).astype(float)

mapping_age_value={'18 - 34 Years':1,'35 - 44 Years':2,'45 - 54 Years':3,'55 - 64 Years':4, '65+ Years':5}
obj_X_train['age_group']=obj_X_train.age_group.map(mapping_age_value)
obj_X_test['age_group']=obj_X_test.age_group.map(mapping_age_value)

mapping_education={'< 12 Years':1,'12 Years':2,'Some College':3,'College Graduate':4}
obj_X_train['education']=obj_X_train.education.map(mapping_education)
obj_X_test['education']=obj_X_test.education.map(mapping_education)

mapping_income={'Below Poverty':1,'<= $75,000, Above Poverty':2,'> $75,000':3}
obj_X_train['income_poverty']=obj_X_train.income_poverty.map(mapping_income)
obj_X_test['income_poverty']=obj_X_test.income_poverty.map(mapping_income)

#mapping_own={'Rent':1,'Own':2}
#obj_X_train['rent_or_own'] = obj_X_train.rent_or_own.map(mapping_own)
#obj_X_train
#obj_X_train[obj_X_train.isnull().any(axis=1)]
```
